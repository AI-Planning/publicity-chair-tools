```python

def generate_organization_committee():
    
    with open('organization-committee', 'r') as f:
        lines = f.readlines()
    
    committee = {}
    positions = []
    
    while lines:
        
        # Fetch the committee position
        position = lines.pop(0).strip()
        positions.append(position)
        committee[position] = []
        
        # Iterate over the members for this position
        line = lines.pop(0).strip()
        while line != '':
            
            # Assumes the following format (all elements required):
            # <website url (no spaces)> <name> (<institution>)
            site = line.split()[0]
            name = ' '.join(line.split('(')[0].split()[1:])
            inst = line.split('(')[-1].split(')')[0]
            committee[position].append((name, inst, site))
            
            if lines:
                line = lines.pop(0).strip()
            else:
                line = ''
    
    # Generate the output
    toRet = ''
    for position in positions:
        
        # Format the position
        toRet += "                                                <h3>%s</h3>\n" % position
        toRet += "                                                <ul>\n"
        
        # Format the members
        for (name, inst, site) in committee[position]:
            toRet += "                                                    <li><a href=\"%s\">%s</a> (%s)</li>\n" % (site, name, inst)
        toRet += "                                                </ul>\n"
    
    return toRet

print generate_organization_committee()
```
