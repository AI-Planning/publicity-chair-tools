```python

# http://www.webtoolhub.com/tn561392-text-to-html-converter.aspx

import csv

SUBS = {
    'Boat': 'Banquet',
    'Complexity (session 11b)': 'Complexity (Session 11b)',
    'Business and Manufacturing Applications (session 6b)': 'Business and Manufacturing Applications (Session 6b)'}
SESSIONS = ['9:35','11:10','13:50','15:20','9:35','11:10','13:50','9:35','11:10','13:50','15:20']
TYPES = {
    'Short': ' (<span class="paper-type">short</span>)',
    'A': ' (<span class="paper-type">application</span>)',
    'R': ' (<span class="paper-type">robotics</span>)',
    'Short J': ' (<span class="paper-type">journal</span>)',
    'Short R': ' (<span class="paper-type">short/robotics</span>)',
    'Short A': ' (<span class="paper-type">short/application</span>)',
    'Wed': '',
    '': ''
}
INDENT = '\n                                        '
html_text = INDENT+'<!-- The following schedule is automatically generated. Do not edit. -->'

def subs(orig):
    return SUBS.get(orig, orig)

with open('ICAPSSchedule-Main.csv', 'r') as f:
    reader = csv.reader(f, delimiter=',')
    main_lines = [[s.strip() for s in row] for row in reader]

with open('ICAPSSchedule-Detailed.csv', 'r') as f:
    reader = csv.reader(f, delimiter=',')
    detailed_lines = [[s.strip() for s in row] for row in reader]

for row in detailed_lines:
    if 'Wednesday' == row[1]:
        html_text += INDENT+"<h3>Wednesday (June 15, 2016)</h3>"
        html_text += INDENT+"<table class=\"schedule\">"
        html_text += INDENT+"  <tr><td class=\"time\">8:20</td><td class=\"opening-remarks\">Opening Remarks</td></tr>"
        html_text += INDENT+"  <tr><td class=\"time\">8:30</td><td class=\"invited-talk\">Invited Talk</td></tr>"
        html_text += INDENT+"  <tr><td class=\"time\">9:30</td><td class=\"break\">Break</td></tr>"
    elif 'Thursday' == row[1]:
        html_text += INDENT+"  <tr><td class=\"time\">16:05</td><td class=\"contest-results\">Competition Results</td></tr>"
        html_text += INDENT+"  <tr><td class=\"time\">16:50</td><td class=\"break\">Break</td></tr>"
        html_text += INDENT+"  <tr><td class=\"time\">17:00</td><td class=\"posters\">Posters &amp; Demos</td></tr>"
        html_text += INDENT+"</table><br />"
        html_text += INDENT+"<h3>Thursday (June 16, 2016)</h3>"
        html_text += INDENT+"<table class=\"schedule\">"
        html_text += INDENT+"  <tr><td class=\"time\">8:30</td><td class=\"invited-talk\">Invited Talk</td></tr>"
        html_text += INDENT+"  <tr><td class=\"time\">9:30</td><td class=\"break\">Break</td></tr>"
    elif 'Friday' == row[1]:
        html_text += INDENT+"  <tr><td class=\"time\">15:20</td><td class=\"break\">Coffee Break</td></tr>"
        html_text += INDENT+"  <tr><td class=\"time\">15:40</td><td class=\"community-meeting\">Community Meeting</td></tr>"
        html_text += INDENT+"  <tr><td class=\"time\">17:40</td><td class=\"banquet\">Conference Banquet</td></tr>"
        html_text += INDENT+"</table><br />"
        html_text += INDENT+"<h3>Friday (June 17, 2016)</h3>"
        html_text += INDENT+"<table class=\"schedule\">"
        html_text += INDENT+"  <tr><td class=\"time\">8:30</td><td class=\"invited-talk\">Invited Talk</td></tr>"
        html_text += INDENT+"  <tr><td class=\"time\">9:30</td><td class=\"break\">Break</td></tr>"
    elif '' == row[1]:
        pass
    elif 'Session' in row[1]:
        if 'Session 2' in row[1]:
            html_text += INDENT+"  <tr><td class=\"time\">10:50</td><td class=\"break\">Coffee Break</td></tr>"
        if 'Session 3' in row[1]:
            html_text += INDENT+"  <tr><td class=\"time\">12:20</td><td class=\"lunch\">Lunch</td></tr>"
        if 'Session 4' in row[1]:
            html_text += INDENT+"  <tr><td class=\"time\">15:00</td><td class=\"break\">Coffee Break</td></tr>"
        if 'Session 6' in row[1]:
            html_text += INDENT+"  <tr><td class=\"time\">10:50</td><td class=\"break\">Coffee Break</td></tr>"
        if 'Session 7' in row[1]:
            html_text += INDENT+"  <tr><td class=\"time\">12:20</td><td class=\"lunch\">Lunch</td></tr>"
        if 'Session 9' in row[1]:
            html_text += INDENT+"  <tr><td class=\"time\">10:50</td><td class=\"break\">Coffee Break</td></tr>"
        if 'Session 10' in row[1]:
            html_text += INDENT+"  <tr><td class=\"time\">12:20</td><td class=\"lunch\">Lunch</td></tr>"
        if 'Session 11' in row[1]:
            html_text += INDENT+"  <tr><td class=\"time\">15:00</td><td class=\"break\">Coffee Break</td></tr>"
        
        if '' == row[5]:
            html_text += INDENT+"  <tr><td class=\"time\">%s</td><td class=\"session-title session-talk\">%s</td></tr>" % (subs(SESSIONS.pop(0)),
                                                                                       subs(row[1]))
        else:
            html_text += INDENT+"  <tr><td class=\"time\">%s</td><td class=\"session-title session-talk1\">%s</td><td class=\"session-title session-talk2\">%s</td></tr>" % (subs(SESSIONS.pop(0)),
                                                                                                     subs(row[1]),
                                                                                                     subs(row[5]))
    else:
        paper1 = "<span class=\"paper-title\">%s%s</div><br />" % (subs(row[1]), TYPES[row[0]])
        paper1 += "<span class=\"paper-author\">%s</div>" % subs(row[2])
        assert '' != row[2]
        if '' == row[5]:
            html_text += INDENT+"  <tr><td class=\"notime\"></td><td class=\"session-talk\">"+paper1+"</td></tr>"
        else:
            paper2 = "<span class=\"paper-title\">%s%s</div><br />" % (subs(row[5]), TYPES[row[4]])
            paper2 += "<span class=\"paper-author\">%s</div>" % subs(row[6])
            assert '' != row[6]
            html_text += INDENT+"  <tr><td class=\"notime\"></td><td class=\"session-talk1\">%s</td><td class=\"session-talk2\">%s</td></tr>" % (paper1, paper2)

html_text += INDENT+"  <tr class=\"conference-end\"><td class=\"time\">16:50</td><td>Closing Remarks</td></tr>"
html_text += INDENT+"</table>"
#with open('schedule.html', 'w') as f:
#    f.write(html_text)

print "Done!"
```
